## Task 1 - Named Entity Recognition

Imports

```python
from nltk.tokenize import word_tokenize
```

Function to Extract the Name Entities

```python
def extract_name_entities(quote):
     words = word_tokenize(quote)
     tagged = nltk.pos_tag(words)
     tree = nltk.ne_chunk(tagged, binary=True)
     return set(
         " ".join(i[0] for i in t)
         for t in tree
         if hasattr(t, "label") and t.label() == "NE"
     )
```

Extracting Name Entities from Sample Text

```python
extract_name_entities(quote)
```

## Task 2 - Making a Frequecy Distribution

Imports

```python
from nltk import FreqDist
```

Frequency Distribution of Twenty Most Occuring Words

```python
frequency_distribution = FreqDist(word_tokenize(quote))
frequency_distribution.most_common(20)
```

Plots of Frequency Distribution of Twenty Most Occuring Words

```python
frequency_distribution.plot(20)
```

```python
frequency_distribution.plot(20, cumulative=True)
```
