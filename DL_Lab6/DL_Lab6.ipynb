### Task 1 - Tokenization

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
```

String to Words

```python
example_string = "Muad'Dib learned rapidly because his first training was in how to learn. And the first lesson of all was the basic trust that he could learn. It's shocking to find how many people do not believe they can learn, and how many more believe learning to be difficult."
words = word_tokenize(example_string)
print(words)
```

```python
words[0]
```

String to Sentences

```python
sentences = sent_tokenize(example_string)
print(sentences)
```

```python
sentences[1]
```

### Task 2 - Removing Stop Words

```python
from nltk.corpus import stopwords
```

Removing Stop Words from a String

```python
word_quote = "Sir, I protest. I am not a merry man!"
stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(word_quote)
  
filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]

print(filtered_sentence)
```

### Task 3 - Stemming

```python
from nltk.stem import PorterStemmer
```

Stemming Words in a String

```python
ps = PorterStemmer()

string_for_stemming = "The crew of the USS Discovery discovered many discoveries. Discovering is what explorers do."
word_tokens = word_tokenize(string_for_stemming)

stemmed_words = [ps.stem(w) for w in word_tokens]

print(stemmed_words)
```

### Task 4 - Tagging Parts of Speech

```python
from nltk import pos_tag
```

Parts of Speech Tagging

```python
sagan_quote = "If you wish to make an apple pie from scratch, you must first invent the universe."
word_tokens = word_tokenize(sagan_quote)

tagged_words = pos_tag(word_tokens)

print(tagged_words)
```

### Task 5 - Lemmatizing

```python
from nltk.stem import WordNetLemmatizer
```

Lemmatizing Words in a String

```python
lmt = WordNetLemmatizer()

string_for_lemmatizing = "The friends of DeSoto love scarves."
word_tokens = word_tokenize(string_for_lemmatizing)

lemmatized_words = [lmt.lemmatize(w) for w in word_tokens]

print(lemmatized_words)
```

### Task 6 - Chunking

```python
from nltk import RegexpParser
```

Chunking Words in a String

```python
lotr_quote = "It's a dangerous business, Frodo; going out your door."
word_tokens = word_tokenize(lotr_quote)
tagged_words = pos_tag(word_tokens)

grammar = "NP: {<DT>?<JJ>*<NN>}"

chunk_parser = RegexpParser(grammar)
chunks = chunk_parser.parse(tagged_words)

print(chunks)
display(chunks)
```
